```python
import numpy as np
from PIL import Image
import PIL, sys
```

```python
center_x = 1080/2
center_y = 720/2

half_width = 500
half_height = 300
```

```python
newsize = (300, 200) 
```

```python
i = 0
fn = 'images/_{}_dfvo_ddppo_nonoise{}'.format(i,'.png')
imageObject = Image.open(fn)
imageObject.crop((center_x - half_width,
                                center_y - half_height, 
                                center_x + half_width,
                                center_y + half_height,
                                )).resize(newsize) 
```

```python
i = 0
for i in range(0,10):
    fn = 'images/_{}_dfvo_ddppo_nonoise{}'.format(i,'.png')
    imageObject = Image.open(fn)
    cropped = imageObject.crop((center_x - half_width,
                                    center_y - half_height, 
                                    center_x + half_width,
                                    center_y + half_height,
                                    ))
    cropped.save(fn, 'PNG')
```
