```python
if __name__ == '__main__':
    num_cpu = 6
    n_steps = 30
    n_updates = 20
    log_dir = './tensorboard'
    config = habitat.get_config("./configs/pointnav_kozub.yaml")
    #
    env = SubprocVecEnv([lambda : environments.NavRLEnvLocalPolicy(config) for i in range(num_cpu)])
    print('environments have been created.')

#     from common import env_utils
#     env = env_utils.construct_envs(habitat.get_config("./configs/pointnav_kozub.yaml"), 
#                                    environments.NavRLEnvLocalPolicy, num_cpu)
#     env.action_space = env.action_spaces
#     env.observation_space = env.observation_spaces


    model = PPO2(policy=LocalPolicy, 
                 env=env,
                 gamma=0.99,
                 cliprange = 0.3, #if wasn't mensioned in ANM paper.
                 verbose=1,
                 n_steps=n_steps,
                 learning_rate=25e-6, #25e-6,
                 ent_coef=1e-3,
                 vf_coef=0.5,
                 tensorboard_log=log_dir,
                 policy_kwargs = {'rgb_shape': [config.SIMULATOR.RGB_SENSOR.HEIGHT, 
                                                config.SIMULATOR.RGB_SENSOR.WIDTH, 3]}
                 )
    model.learn(total_timesteps=num_cpu*n_steps*n_updates)
    print('learning complete!')
```

```python
model_path = './LP_PPO_model'
model.save(save_path =model_path)
del model
```

```python
import common.default_blocks as db
config = habitat.get_config("./configs/pointnav_kozub.yaml")
model_path = './LP_PPO_model'

env = environments.NavRLEnv(config)
#env = SubprocVecEnv([lambda: environments.NavRLEnv(config) for i in range(6)])
obs = env.reset()
agent = db.ApenAIPPOBlockAgent(config.ENVIRONMENT.MAPSCALE, model_path, 25)

counter = 0
while True:
    action = agent.act(env, obs)
    obs, reward, done, info = env.step(action+1)
    print(counter, '- step /', action, ' - action /', reward, '- rew /', 
                      obs['pointgoal_with_gps_compass'] , '- global goal / ', info, '- metric')
    counter += 1
    if done:
        break
print(counter)

env.close()
```
